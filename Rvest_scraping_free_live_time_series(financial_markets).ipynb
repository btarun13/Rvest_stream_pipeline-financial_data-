## Using Rvest on investing dotcom for scraping information

### Following libraries are used for getting and cleaning the structured data.
### We specify the url we want to scrape data from then check if scraping is allowed. Note some websites don't allow scraping so it saves time if you check it before hand

```r
install.packages("robotstxt")
install.packages("tidyverse")
install.packages("rvest")
install.packages("stringr")
```

```r
library(robotstxt)
library(tidyverse)
library(rvest)
library(stringr)
url4 <- "https://uk.investing.com/currencies/streaming-forex-rates-majors"
paths_allowed("https://uk.investing.com/currencies/streaming-forex-rates-majors")

Sys.Date()
i = 1
```

### as the output is true I can scrape data

```r
url4 <- "https://uk.investing.com/currencies/streaming-forex-rates-majors"  ## location of scrape
yourdata <- url4 %>% read_html() %>% 
html_nodes(xpath = '//*[@id="cross_rates_container"]/table') %>%   ## this is something you should note by inspect, after that read the id and put it in here
html_table()   

yourdata <- as.data.frame(yourdata)

data_snap<- yourdata[,2:9]
data_snap$Date <- Sys.Date()
```

## Code for storing streamed data, all data collected for free without any commercial APIs

```r
while ( TRUE ){ ## endless loop 
  Sys.sleep(0.5)   ## taking snapshot evert 0.5 secs
  url4 <- "https://uk.investing.com/currencies/streaming-forex-rates-majors"  ## location of scrape
  yourdata <- url4 %>% read_html() %>% 
    html_nodes(xpath = '//*[@id="cross_rates_container"]/table') %>%   ## this is something you should note by inspect, after that read the id and put it in here
    html_table()   
  
  yourdata <- as.data.frame(yourdata)
  
  data_snap<- yourdata[,2:9]
  data_snap$Date <- Sys.Date()
  if (i == 1){
    write.table(data_snap, file = '/home/tarun/Music/stream_forex.csv',append = T,row.names = F,col.names = T)  ## put dataframe with col name
  } else{
    write.table(data_snap, file = '/home/tarun/Music/stream_forex.csv',append = T,row.names = F,col.names = F)   ## after first this will only append existing dataset
  }
  url4 <- "https://uk.investing.com/commodities/real-time-futures"
  yourdata <- url4 %>% read_html() %>% 
    html_nodes(xpath = '//*[@id="cross_rates_container"]/table') %>% 
    html_table()   
  
  yourdata <- as.data.frame(yourdata)
  
  data_snap<- yourdata[,2:9]
  data_snap$Date <- Sys.Date()
  if (i == 1){
    write.table(data_snap, file = '/home/tarun/Music/stream_commodities.csv',append = T,row.names = F,col.names = T)
  } else{
    write.table(data_snap, file = '/home/tarun/Music/stream_commodities.csv',append = T,row.names = F,col.names = F)
  }
  i <- i+1
}
```
